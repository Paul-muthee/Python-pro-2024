## Boxplots

```python
import pandas as pd
import numpy as np
import plotly.express as px

# Sample data
df = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=200),
    'Value': np.random.randn(200)
})

# Box plot
fig = px.box(df, x='Category', y='Value', title='Box Plot by Category')
fig.show()
```

## 3d Surface

```python
import numpy as np
import plotly.graph_objects as go

# Create data for surface plot
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))  # example surface

# Surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot', scene=dict(
    xaxis_title='X',
    yaxis_title='Y',
    zaxis_title='Z'
))
fig.show()
```

## Spread

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Generate data
np.random.seed(42)
x = np.arange(0, 100)
y_mean = np.sin(x * 0.1)
y_upper = y_mean + 0.2 + np.random.rand(100) * 0.1
y_lower = y_mean - 0.2 - np.random.rand(100) * 0.1

# Plot spread between y_upper and y_lower
fig = go.Figure([
    go.Scatter(x=x, y=y_upper, line=dict(color='lightblue'), name='Upper Bound'),
    go.Scatter(x=x, y=y_lower, fill='tonexty', line=dict(color='lightblue'), name='Lower Bound', fillcolor='rgba(173,216,230,0.4)'),
    go.Scatter(x=x, y=y_mean, line=dict(color='blue'), name='Mean')
])

fig.update_layout(title='Spread Plot (Shaded Area Between Bounds)', xaxis_title='X', yaxis_title='Y')
fig.show()
```
