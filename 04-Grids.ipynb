# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
g = sns.pairplot(iris)
for ax in g.axes.flatten():
    ax.set_facecolor('#f0f0f0') 
    ax.grid(True, color='gray', linewidth=0.5, alpha=0.9)
plt.show()
```

## PairGrid

Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
sns.PairGrid(iris)
sns.set_theme(style="darkgrid")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
iris = sns.load_dataset("iris")

# Create PairGrid
g = sns.PairGrid(iris)
g.map(plt.scatter)

# Customize each subplot
for ax in g.axes.flatten():
    ax.set_facecolor('#f0f0f0')               # Light grey background
    ax.grid(True, color='gray', linewidth=0.5, alpha=0.7)  # Thin, gray grid

# Optional: match the outer figure background
g.fig.set_facecolor('#e6e6e6')

plt.show()
```

```python
# Then you map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter)
sns.set_theme(style="darkgrid")
plt.show()
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
sns.set_theme(style="darkgrid")
plt.show()
```

## pairplot

pairplot is a simpler version of PairGrid (you'll use quite often)

```python
sns.pairplot(iris)
sns.set_theme(style="darkgrid")
plt.show()
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
sns.set_theme(style="darkgrid")
plt.show()
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Just the Grid
sns.set_theme(style="darkgrid")
g = sns.FacetGrid(tips, col="time", row="smoker")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
sns.set_theme(style="darkgrid")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Notice hwo the arguments come after plt.scatter call
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
sns.set_theme(style="darkgrid")
plt.show()
```

## JointGrid

JointGrid is the general version for jointplot() type grids, for a quick example:

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```
