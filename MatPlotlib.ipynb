# Setting other properties like line color,tranparency, and more

```python
fig,axes=plt.subplots(dpi=150)
axes.plot(x, x+1,color="red",alpha=0.5)#aplpha makes it blurred or light
axes.plot(x, x+2, color="blue")
axes.plot(x, x+3, color="green")
```

```python
fig,ax=plt.subplots(dpi=100)
ax.plot(x,x+1, color="blue",linewidth=0.25)
ax.plot(x,x+2, color="blue",linewidth=1)
ax.plot(x,x+3, color="blue",linewidth=1.5)
ax.plot(x,x+4, color="blue",linewidth=2)
```

```python
fig,ax=plt.subplots(dpi=100)
ax.plot(x,x+1, color="blue",lw=2,linestyle='-')# we can use lw for linewidth
ax.plot(x,x+2, color="blue",lw=2, linestyle='-.')
ax.plot(x,x+3, color="blue",lw=2,linestyle=':')
line, =ax.plot(x,x+4, color="black",lw=2,)
line.set_dashes([5,10,15,10])
```

```python
fig,axes=plt.subplots(dpi=150)
axes.plot(x, x+1,color="red", marker='o', markersize=4, markerfacecolor='blue')#markersize increases the size and markerfacecolor is that of shapes
axes.plot(x, x+2, color="blue",marker='+')
axes.plot(x, x+3, color="green",marker='s')
```

# Setting canvas grid and axis range

```python
fig, Axes = plt.subplots(1, 2, figsize=(10, 5))

Axes[0].plot(x,x**2, x,x**3, lw=2)
Axes[0].grid(True)

Axes[1].plot(x,x**2, x,x**3, lw=2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
Axes[1].grid(True)


```

# Other 2D Graphics

```python
n=np.array([0,1,2,3,4,5])
fig, axes=plt.subplots(1,4, figsize=(16,5))
axes[0].set_title('scatter')
axes[0].scatter(x, x+0.25*np.random.randn(len(x)))

axes[1].set_title('step')
axes[1].step(n,n**2,lw=2)

axes[2].set_title('bar')
axes[2].bar(n, n**2,)

axes[3].set_title('fill_between')
axes[3].fill_between(x, x**2,x**3, color="green")
```

```python
#Draw Histogram
n=np.random.randn(100000)
fig, axes= plt.subplots(1,2, figsize=(12,4))
axes[0].set_title("Default Histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative detailed Histogram")
axes[1].hist(n, cumulative=True, bins=10)
```

# Contour Maps

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0, delta)
X,Y= np.meshgrid(x,y)
Z1=np.exp(-X**2-Y**2)
Z2=np.exp(-(X-1)**2-(Y-1)**2)
Z=(Z1-Z2)*2
print(X)
print(Y)
```

```python
fig, ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('contour title')
```
