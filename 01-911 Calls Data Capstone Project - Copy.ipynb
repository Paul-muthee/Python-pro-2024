** Now see if you can use seaborn's lmplot() to create a linear fit on the number of calls per month. Keep in mind you may need to reset the index to a column. **

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Group by month and count calls
byMonth = df.groupby('Month').count()

# Step 2: Reset the index so 'Month' becomes a column
byMonth = byMonth.reset_index()

# Step 3: Use lmplot to plot a linear regression on number of calls per month
sns.lmplot(x='Month', y='title', data=byMonth)

# Add labels and title
plt.title('Linear Fit of 911 Calls per Month')
plt.xlabel('Month')
plt.ylabel('Number of Calls')
plt.tight_layout()

plt.show()
```

** Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
import matplotlib.pyplot as plt

# Set figure size for subplots
plt.figure(figsize=(18, 12))

# Define the unique reasons
reasons = ['EMS', 'Fire', 'Traffic']

# Loop through each reason and create a subplot
for i, reason in enumerate(reasons, 1):
    # Filter DataFrame by reason
    df_reason = df[df['Reason'] == reason]
    
    # Group by 'Date' and count number of calls
    calls_by_date = df_reason.groupby('Date').count()
    
    # Plotting
    plt.subplot(3, 1, i)
    calls_by_date['title'].plot(color='C' + str(i))
    plt.title(f'911 Calls per Day - {reason}')
    plt.xlabel('Date')
    plt.ylabel('Number of Calls')
    plt.tight_layout()
    plt.grid(True)

# Show all plots
plt.show()
```

** Now create a HeatMap using this new DataFrame. **

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure timeStamp is datetime type
df['timeStamp'] = pd.to_datetime(df['timeStamp'])

# Create new columns for hour and day of week
df['Hour'] = df['timeStamp'].dt.hour
df['DayOfWeek'] = df['timeStamp'].dt.dayofweek  # Monday=0, Sunday=6

# Optional: Map numeric days to labels
day_map = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
df['DayOfWeek'] = df['DayOfWeek'].map(day_map)

# Create pivot table: rows = day of week, columns = hour, values = call counts
heatmap_data = df.groupby(['DayOfWeek', 'Hour']).size().unstack()

# Reorder rows to start from Monday
ordered_days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
heatmap_data = heatmap_data.reindex(ordered_days)

# Plot heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5, linecolor='gray')

plt.title('911 Calls by Day of Week and Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.tight_layout()
plt.show()
```

** Now create a clustermap using this DataFrame. **

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Ensure timeStamp is datetime
df['timeStamp'] = pd.to_datetime(df['timeStamp'])

# Extract hour and day of week
df['Hour'] = df['timeStamp'].dt.hour
df['DayOfWeek'] = df['timeStamp'].dt.dayofweek  # Monday=0, Sunday=6

# Optional: map to day names
day_map = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
df['DayOfWeek'] = df['DayOfWeek'].map(day_map)

# Create pivot table
clustermap_data = df.groupby(['DayOfWeek', 'Hour']).size().unstack()

# Reorder the rows for logical weekday order
ordered_days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
clustermap_data = clustermap_data.reindex(ordered_days)

# Plot clustermap
sns.clustermap(clustermap_data, cmap='viridis', linewidths=0.5, figsize=(10, 6), standard_scale=1)

plt.suptitle('911 Calls Clustermap by Day of Week and Hour of Day', y=1.05)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure timeStamp is datetime
df['timeStamp'] = pd.to_datetime(df['timeStamp'])

# Extract Day of Week and Month
df['DayOfWeek'] = df['timeStamp'].dt.dayofweek  # 0 = Mon, 6 = Sun
df['Month'] = df['timeStamp'].dt.month  # 1 = Jan, 12 = Dec

# Map to names
day_map = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
             7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}

df['DayOfWeek'] = df['DayOfWeek'].map(day_map)
df['Month'] = df['Month'].map(month_map)
```

```python
# Pivot table: Rows = DayOfWeek, Columns = Month, Values = call counts
month_pivot = df.groupby(['DayOfWeek', 'Month']).size().unstack()

# Reorder for readability
ordered_days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
ordered_months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                  'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

month_pivot = month_pivot.reindex(index=ordered_days, columns=ordered_months)
plt.figure(figsize=(14, 6))
sns.heatmap(month_pivot, cmap='YlOrRd', linewidths=0.5, linecolor='gray', annot=True, fmt='g')

plt.title('911 Calls Heatmap by Day of Week and Month')
plt.xlabel('Month')
plt.ylabel('Day of Week')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure datetime
df['timeStamp'] = pd.to_datetime(df['timeStamp'])

# Extract features
df['DayOfWeek'] = df['timeStamp'].dt.dayofweek
df['Month'] = df['timeStamp'].dt.month

# Map to names
day_map = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
             7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}

df['DayOfWeek'] = df['DayOfWeek'].map(day_map)
df['Month'] = df['Month'].map(month_map)

# Create pivot table
month_pivot = df.groupby(['DayOfWeek', 'Month']).size().unstack()

# Reorder rows and columns
ordered_days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
ordered_months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                  'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_pivot = month_pivot.reindex(index=ordered_days, columns=ordered_months)

# ðŸ”§ Fill NaNs with 0
month_pivot = month_pivot.fillna(0)

# Plot clustermap
sns.clustermap(month_pivot, cmap='coolwarm', linewidths=0.5,
               figsize=(12, 8), standard_scale=1)

plt.suptitle('911 Calls Clustermap by Day of Week and Month', y=1.05)
plt.show()
```
